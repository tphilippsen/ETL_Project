```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# import first data set
pd.set_option('max_colwidth', 200)
Pop_df = pd.read_csv("Population_data.csv")
```

```python
# Sanity check to make sure CSV loaded
Pop_df.head()
```

```python
# Filter data by population total 
population_total = Pop_df.loc[(Pop_df['Series Code'] == 'SP.POP.TOTL')]
population_total
```

```python
# Selecting target columns
clean_pop_df = population_total[['Series Name', 'Country Code', '2010 [YR2010]', '2011 [YR2011]',
       '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]']].copy()

clean_pop_df.head()
```

```python
# Rename columns
renamed_pop_df = clean_pop_df.rename(columns={'2010 [YR2010]': '2010', '2011 [YR2011]': '2011',
       '2012 [YR2012]': '2012', '2013 [YR2013]': '2013', '2014 [YR2014]': '2014', '2015 [YR2015]': '2015'})
renamed_pop_df.head()
```

```python
# import second data set
pd.set_option('max_colwidth', 200)
gdp_df = pd.read_excel('API_NY.GDP.MKTP.CD_DS2_en_excel_v2_4770502.xls', header=3)
```

```python
# Sanity check to make sure XLS loaded
gdp_df.head()
```

```python
# Identify columns
gdp_df.columns
```

```python
# Selecting target columns
clean_gdp_df = gdp_df[['Country Name', 'Country Code', 'Indicator Name','2010', '2011', '2012', '2013', '2014', '2015']].copy()

clean_gdp_df.head()
```

```python
# Renaming indicator column
clean_gdp_df = clean_gdp_df.rename(columns={'Indicator Name': 'Indicator'})
clean_gdp_df
```

```python
# Identifying columns
clean_gdp_df.columns
```

```python
# Merge dataframes on country code
gdp_pop_df = pd.merge(clean_gdp_df, renamed_pop_df, on= 'Country Code', how= 'right')
gdp_pop_df
```

```python
# Identify columns
gdp_pop_df.columns
```

```python
# Renaming columns
rename_gdp_pop_df = gdp_pop_df.rename(columns={'Country Name': 'Country','Indicator Name': 'Indicator', '2010_x': 'GDP_2010', '2011_x': 'GDP_2011',
       '2012_x': ' GDP_2012', '2013_x': 'GDP_2013', '2014_x': 'GDP_2014', '2015_x': 'GDP_2015', '2010_y': 'POP_2010',
       '2011_y': 'POP_2011', '2012_y': 'POP_2012', '2013_y': 'POP_2013', '2014_y': 'POP_2014', '2015_y': 'POP_2015'})
```

```python
# Print database
rename_gdp_pop_df
```

```python
# Drop N/A from rows and columns
rename_gdp_pop_df.dropna()
rename_gdp_pop_df
```

```python
# Final dataset
rename_gdp_pop_df
```

```python
# Export to csv
rename_gdp_pop_df.to_csv('GDP_Population_Data_By_Country.csv')
```
